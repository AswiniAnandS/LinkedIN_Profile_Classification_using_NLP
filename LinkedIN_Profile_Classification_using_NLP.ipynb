The unnecessary information like URLS, hashtags and other non necessary characters were removed and Cleansed from the "combined" column

```python
import re
def clean_data(Text):
    Text = re.sub('http\S+\s*', ' ', Text)  # remove URLs
    Text = re.sub('RT', ' ', Text)  # remove RT
    Text = re.sub('#\S+', '', Text)  # remove hashtags
    Text = re.sub('@\S+', '  ', Text)  # remove mentions
    Text = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', Text)  # remove punctuations
    Text = re.sub(r'[^\x00-\x7f]',r' ', Text)
    Text = re.sub('\s+', ' ', Text)  # remove extra whitespace
    return Text

df1['cleandata'] = df1.Combined.apply(lambda x: clean_data(x))
```

Creating a corpus for skills

```python
import nltk
nltk.download('popular', quiet=True)
#nltk.download('punkt')

from nltk.corpus import stopwords
import string
from wordcloud import wordcloud

stop_words = set(stopwords.words('english')+['``',"''","'",""])


bagofwords=[]
df_info=df1['cleandata'].values
sentence=""
for i in df_info:
    sentence+=i
    tokenwords=nltk.word_tokenize(i)
    for word in tokenwords:
        if word not in stop_words and word not in string.punctuation:
          bagofwords.append(word)
```

Creating a wordcount and displaying the common words along with its frequency

```python
wordcount = nltk.FreqDist(bagofwords)
commonwords = wordcount.most_common(50)
print(commonwords)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack

ip_data = df2['cleandata'].values
tgt = df2['category'].values

word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    stop_words='english')
word_vectorizer.fit(ip_data)
WordFeatures = word_vectorizer.transform(ip_data)

print ("Feature completed .....")

X_train,X_test,y_train,y_test = train_test_split(WordFeatures,tgt,random_state=42, test_size=0.3,
                                                 shuffle=True, stratify=tgt)
print(X_train.shape)
print(X_test.shape)
print(WordFeatures.shape)
```

CONFUSION MATRIX

A confusion matrix is a summary of prediction results on a classification problem.

Positive (P) : Observation is positive
Negative (N) : Observation is not positive
True Positive (TP) : Observation is positive, and is predicted to be positive.
False Negative (FN) : Observation is positive, but is predicted negative.
True Negative (TN) : Observation is negative, and is predicted to be negative.
False Positive (FP) : Observation is negative, but is predicted positive.

The confusion matrix helps us to identify how many of the classifier model's prediction were correct and how many were wrong. The rows are the true labels and the columns are the predicted labels.

The numbers in the diagonal is the number of times the predicted value matches the true value.

The numbers in the other cells represents the number of times the classifier predicted incorrectly.
